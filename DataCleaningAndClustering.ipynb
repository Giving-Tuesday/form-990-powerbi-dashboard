```python
import warnings

import pandas as pd

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"Sample_Program_Service_Data.csv", index_col=0, encoding='latin')
```

```python
df2 = pd.DataFrame()
df2["UID"] = df["EIN"].map(str) + "_" + df["FISYR"].map(str)
df2["EIN"]  = df["EIN"] 
df2["FISYR"]  = df["FISYR"]
df2["NteeFinal"] = df["NteeFinal"]
df2["NteeCategory"] = df["NteeFinal"].map(lambda x: str(x)[0])
df2["MISSION"] = df["MISSION"]
df2["proj1Desc"] = df["proj1Desc"].fillna(" ")
df2["proj2Desc"] = df["proj2Desc"].fillna(" ")
df2["proj3Desc"] = df["proj3Desc"].fillna(" ")
df2["ProjAllText"] = df2["proj1Desc"] + ". " + df2["proj2Desc"] + ". " + df2["proj3Desc"]
df2["ProjAllText"] = df2["ProjAllText"].fillna(" ").map(lambda x: str(x).replace("\t", " "))
df2["TOTALREVCURRENT"] = df["TOTALREVCURRENT"].fillna(0)
df2["TOTALEXPCURRENT"] = df["TOTALEXPCURRENT"].fillna(0)
df2["TOTALEXPCURRENT"] = df["TOTALEXPCURRENT"].fillna(0)
df2["DISCOPS"] = df["DISCOPS"] == "X"
df2["TOTEMPLOYEE"] = df["TOTEMPLOYEE"].fillna(" ")
df2["TOTVOLUNTEERS"] = df["TOTVOLUNTEERS"].fillna(" ")
df2["STATE"] = df["STATE"].fillna(" ")
```

```python
df2.groupby("DISCOPS").head()
```

```python
def NteeCategoryName(x):
    if x == "A":
        return "Arts, Culture, and Humanities"
    elif x == "B":
        return "Education"
    elif x in ["C","D"]:
        return "Environment and Animals"
    elif x in ["E", "F", "G", "H"]:
        return "Health"
    elif x in ["I", "J", "K", "L", "M", "N", "O", "P"]:
        return "Human Services" 
    elif x in ["Q"]:
        return "International, Foreign Affairs" 
    elif x in ["R", "S", "T", "U", "V", "W"]:
        return "Public, Societal Benefit"
    elif x in ["X"]:
        return "Religion Related" 
    elif x in ["Y"]:
        return "Mutual/Membership Benefit" 
    elif x in ["Z"]:
        return "Unknown, Unclassified" 
    else:
        return "Code Not In Mapping" 
df2["NteeCategoryName"] = df2["NteeCategory"].map(lambda x: NteeCategoryName(x))
```

```python
df2["WEBSITE"] = df["WEBSITE"]
df2["HasWebSite"] = df["WEBSITE"].isnull().map(lambda x : not x)
df2.set_index("UID")
```
